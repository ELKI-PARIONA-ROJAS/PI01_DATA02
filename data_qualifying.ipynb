```python
# importar la data del archivo csv y convertirlo en un dataframe
import pandas as pd
import numpy as np

data_qua1 = pd.read_json('Datasets/Qualifying/qualifying_split_1.json')
data_qua2 = pd.read_json('Datasets/Qualifying/qualifying_split_2.json')

data_qua = pd.concat([data_qua1, data_qua2], ignore_index=True)
```

```python
data_qua.info()
```

```python
# Cambiar el nombre de las columnas
data_qua.rename(columns={'qualifyId':'IdCalificacion','raceId':'IdCarrera','driverId':'IdPiloto',
                        'constructorId':'IdConstructor','number':'Nro_X','position':'Posicion',
                        'q1':'Calificacion1','q2':'Calificacion2','q3':'Calificacion3'}, inplace=True)
```

```python
# Converir IdCalificacion en indice
data_qua.set_index('IdCalificacion', inplace=True)
```

```python
# contar cuantos valores de tipo '\\N' hay en cada columna
data_qua.apply(lambda x: sum(x=='\\N'), axis=0)
```

```python
# Reemplazar los valores '\\N' y los vacios por valores nulos
data_qua = data_qua.replace('\\N', np.nan)
data_qua = data_qua.replace('', np.nan)
```
