```python
# importar las librerias y la base de datos

import pandas as pd
import numpy as np

data_results = pd.read_json('Datasets/results.json', lines=True)
```

```python
data_results.head()
```

```python
data_results.info()
```

```python
# renombrar las columnas
data_results.rename(columns={'resultId':'IdResultado','raceId':'IdCarrera','driverId':'IdPiloto','constructorId':'IdConstructor',
'number':'Nro_X','grid':'Parrilla','position':'Posicion','positionText':'PosicionTexto','positionOrder':'PosicionOrden',
'points':'Puntos','laps':'Vueltas','time':'Tiempo','milliseconds':'Duracion(ms)','fastestLap':'VueltaRapida','rank':'Rango',
'fastestLapTime':'TiempoVueltaRapida','fastestLapSpeed':'VelocidadVueltaRapida','statusId':'IdEstado'}, inplace=True)
```

```python
# Eliminar los duplicados en la columna de IdResultado
data_results.drop_duplicates(subset='IdResultado', keep='first', inplace=True)
```

```python
# establecer a driverId como indice
data_results = data_results.set_index('IdResultado')
```

```python
# contar cuantos valores de tipo '\\N' hay en cada columna
data_results.apply(lambda x: sum(x=='\\N'), axis=0)
```

```python
# Reemplazar los valores '\\N' y los vacios por valores nulos
data_results = data_results.replace('\\N', np.nan)
data_results = data_results.replace('', np.nan)

# Reemplazar los valores '\\N' y nulos por valores vacios
# data_results = data_results.replace('\\N', '')
# data_results = data_results.replace(np.nan, '')
```

```python
# Extraer en otra dataframe los valores nulos los que no tienen nulos en la columna 'Duracion(ms)'
results_notnull = data_results[data_results['Duracion(ms)'].notnull()]
```
